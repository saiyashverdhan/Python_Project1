Q2. Find top 5 players with most valuable contracts

```python
# creating of original dataframe
backup_df = df.copy()
```

```python
df[['Name','Value']]
```

```python
# Data clean - Removing currency sign and replaceing millions and thousands with their zeroes
value_clean = df['Value'].str.replace("€","").str.strip()
value_clean
```

```python
# value_clean.str.replace('M', '000000').str.replace('K', '000')
# 67.5 looses its million value because of the decimal point. Hence, need to deal with decimals as well.
```

```python
# creating a function to assign the multipliers instead of M and K strings
# This function that takes one value at a time and processes it
def multiply(value):
    if 'M' in value:
        return float(value.replace("M","")) * 1000000
    if 'K' in value:
        return float(value.replace("K","")) * 1000
    else:
        return float(value)
```

```python
# # Applying the function to each element in the 'Value' column
pd.options.mode.chained_assignment = None  # suppresses the warning
df['Value(€)'] = value_clean.apply(multiply)
df['Value(€)']
```

```python
q2_df = df.sort_values(by='Value(€)',ascending=False)
q2_df[['Name','Value(€)']].head(5)
```

Q3. Find players with least value at release

```python
q2_df.head(1)
```

```python
# TO clean release clause, performing same operations as Q2
value_clean_q3 = df['Release Clause'].str.replace("€","").str.strip()
value_clean_q3
```

```python
q2_df['Release_Clause(€)'] = value_clean_q3.apply(multiply)
q2_df['Release_Clause(€)']
```

```python
# Finding differnce in release clause and value to get capital profit from the player
q2_df['Least_Value'] = q2_df['Release_Clause(€)'] - q2_df['Value(€)']
q2_df['Least_Value']
```

```python
q3_df = q2_df.sort_values(by='Least_Value',ascending=True)
```

Q4. Which players are highly valuable but still underpaid (on low wages)? (hint: scatter plot between wage and value)

```python
# Applying same cleaning method as q2
# TO clean release clause, performing same operations as Q2
value_clean_q4 = df['Wage'].str.replace("€","").str.strip()
value_clean_q4
```

```python
# Applying the same function
pd.options.mode.chained_assignment = None  # suppresses the warning
df.loc[:,'Wage(€)'] = value_clean_q4.apply(multiply)
df.loc[:,'Wage(€)']
```

```python
q4_df = df[['Wage(€)','Value(€)']]
q4_df
```

```python
# Scatter plot
q4_df.plot.scatter(x= 'Value(€)' , y = 'Wage(€)')
```

```python
# To find the players, creating a metric of underpaid players
pd.options.mode.chained_assignment = None  # suppresses the warning
df['Wage/Value'] = df['Wage(€)'] / df['Value(€)']
df['Wage/Value']
```

```python
#Less ratio would mean an underpaid player
sorted_df = df.sort_values(by='Wage/Value')
```

```python
top_underpaid = sorted_df[['Name', 'Value(€)', 'Wage(€)']].head(5)
top_underpaid.plot.bar(x='Name', y='Value(€)')
```
